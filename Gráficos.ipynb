```python
dados = pd.read_excel("/content/2019volleyball.xlsx")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'Altura (cm)', y = 'Peso (kg)', data = dados, hue = "Nacionalidade", s = 150, style= "Mão Dominante" )
plt.title("Gráfico de Dispersão: Altura(cm) vs Peso (kg)")
```

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(x = 'Salto (cm)', y = 'Peso (kg)', data = dados, style = "Mão Dominante", s = 150, hue = "Mão Dominante")
plt.title("Gráfico dispesão: Salto (cm) vs Peso (kg)", fontsize = 15,)
plt.xlabel("Salto (cm)", fontsize = 15)
plt.ylabel("Peso (cm)", fontsize = 15)
```

```python
dados1 = pd.DataFrame({'Ano' : [2019, 2020, 2021, 2022],
                       'Número' : [2, 11, 55, 23],
                       'Tipo' : ["Azul", "Preto", "Branco", "Rosa"]})
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x = 'Ano', y = 'Número', data = dados1, hue = "Tipo")
plt.title("Gráfico de Barras: Ano vs Número", fontsize =15)
plt.xlabel("Ano", fontsize = 15)
plt.ylabel("Número", fontsize = 15)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(dados["Posição"], order = dados["Posição"].value_counts().index)
plt.xlabel("Posição", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
```

```python
plt.figure(figsize=(10,10))
dados["Peso (kg)"].hist()
```

```python
labels = dados["Nacionalidade"]. unique()
sizes = dados['Nacionalidade'].value_counts().values

#visuals
plt.figure(figsize = (10,10))
plt.pie(sizes, labels = labels, autopct = '%1.1f%%')
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = 'Posição', y = 'Altura (cm)', data = dados)
```

```python
correlation = dados.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, annot = True, cmap = 'coolwarm', vmax=.9, center = 0, square = True, linewidths = 2, fmt = '.1f', cbar_kws = {"shrink": .5})
```
